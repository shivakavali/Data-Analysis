```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("C://Temp/gdp.csv")
```

```python
df
```

```python
df_ind = df[df["Country Name"]=="India"]
df_ind.plot(kind = "line", x = "Year", y = "Value", figsize = (8,6), grid = True)
```

```python
df_ind
```

```python
ind = df_ind.values
lst2 = [0]
for i in range(1, len(ind)):
    pre = ind[i-1][3]
    cur = ind[i][3]
    change = round(((pre-cur)/pre)*100, 2)
    lst2.append(change)
df_ind.assign(GDP = lst2)
```

```python
final_data = []
for country_name in df["Country Name"].unique():
    df_country = df[df["Country Name"] == country_name]
    data = df_country.values
    gdp_change = [0]
    for i in range(1, len(data)):
        prev = data[i-1][3]
        cur = data[i][3]
        dif =  round(((cur-prev)/prev)*100,2)
        gdp_change.append(dif)
    df_country = df_country.assign(GDP=gdp_change)
    final_data.append(df_country)
```

```python
df = pd.concat(final_data, axis = 0)
```
