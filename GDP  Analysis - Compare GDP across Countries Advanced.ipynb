```python
lst1 = []
for c_name in df["Country Name"].unique():
    c_data = df[df["Country Name"]==c_name]
    data = c_data.values
    lst = [0]
    for i in range(1, len(data)):
        prev = data[i-1][3]
        cur = data[i][3]
        dif = round(((cur-prev)/prev)*100,2)
        lst.append(dif)
    c_data = c_data.assign(GDP = lst)
    lst1.append(c_data)
```
